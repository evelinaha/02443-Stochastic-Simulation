## Part 1

```python
#Exercise 13 in Chapter 8 of Ross (p. 152)
n = 10
X = np.array([56,101,78,67,93,87,64,72,80,69])
a = -5
b = 5

reps = 10000

X_mean = np.mean(X)

B_samples = np.random.choice(X,size=(reps, n),replace=True)
B_means = np.mean(B_samples,axis=1)

p_est = np.mean((B_means - X_mean > a) & (B_means - X_mean < b))

print("Data mean:",X_mean)
print("Bootstrap estimate of p:",p_est)
```

## Part 2

```python
#Exercise 15 in Chapter 8 of Ross (p. 152)
n = 15
X = np.array([5,4,9,6,21,17,11,20,7,10,21,15,13,16,8])

reps = 10000

X_var = np.var(X,ddof=1)

B_samples = np.random.choice(X,size=(reps, n),replace=True)
B_vars = np.var(B_samples,axis=1,ddof=1)

B_meanVar = np.mean(B_vars)

B_varVar = np.var(B_vars,ddof=1)

print("Data variance:", X_var)
print("Bootstrap mean of variance:", B_meanVar)
print("Bootstrap variance of variance:",B_varVar)
```

## Part 3

```python
n = 200
beta = 1
k = 1.05
X = beta * (np.random.pareto(k,n) + 1)

reps = 100
```

### 1)

```python
X_mean = np.mean(X)
X_median = np.median(X)

print("Sample mean:", X_mean)
print("Sample median:", X_median)
```

### 2)

```python
B_samples = np.random.choice(X, size=(reps, n), replace=True)

B_means = np.mean(B_samples, axis=1)

B_varMean = np.var(B_means,ddof=1)

print("Bootstrap variance of mean:",B_varMean)
```

### 3)

```python
B_medians = np.median(B_samples, axis=1)

B_varMedian = np.var(B_medians,ddof=1)

print("Bootstrap variance of median:",B_varMedian)
```

### 4)

```python
print("Bootstrap variance of mean vs median:",B_varMean,B_varMedian)

#Comparison: variance of the mean is much larger than that of the median
```

The observed difference is likely due to the Pareto distribution having very extreme values affecting the mean.
